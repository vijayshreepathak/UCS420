```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# 1. Initialize Random Seed
roll_number = 102316022
np.random.seed(roll_number)

# 2. Generate Sales Data
sales_data = np.random.randint(1000, 5001, size=(12, 4))

# 3. Convert to DataFrame
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']
sales_df = pd.DataFrame(sales_data, columns=categories, index=months)


# 1. Display first 5 rows and summary statistics
print(sales_df.head())
print(sales_df.describe())

# 2. Total sales per category and total sales per month
total_sales_category = sales_df.sum()
total_sales_month = sales_df.sum(axis=1)

# 3. Average sales growth between consecutive months for each category
average_growth = sales_df.pct_change().mean()

# 4. Add Total Sales and Growth Rate columns
sales_df['Total Sales'] = total_sales_month
sales_df['Growth Rate'] = sales_df['Total Sales'].pct_change() * 100

# 5. Apply randomized discount
if roll_number % 2 == 0:
    sales_df['Electronics'] *= 0.9
else:
    sales_df['Clothing'] *= 0.85

# Part III: Visualizations (Matplotlib and Seaborn)

# 1. Line plots for monthly sales trends
plt.figure(figsize=(10, 6))
sns.lineplot(data=sales_df[categories])
plt.title('Monthly Sales Trends')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend(categories)
plt.show()

# 2. Box plots for sales distribution
plt.figure(figsize=(8, 6))
sns.boxplot(data=sales_df[categories])
plt.title('Sales Distribution by Category')
plt.show()







```

```python
# Q2: Operations on a given array
array = np.array([[1, -2, 3], [-4, 5, -6]])

# i. Element-wise absolute value
absolute_values = np.abs(array)
print(absolute_values)

# ii. Percentiles
percentiles = {
    '25th': np.percentile(array, 25, axis=None),
    '50th': np.percentile(array, 50, axis=None),
    '75th': np.percentile(array, 75, axis=None)
}
print(percentiles)

# iii. Mean, Median, Standard Deviation
stats = {
    'Mean': np.mean(array),
    'Median': np.median(array),
    'Standard Deviation': np.std(array)
}
print(stats)
```

```python
# Q3: Floor, ceiling, truncated, rounded values
a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

floor_values = np.floor(a)
ceiling_values = np.ceil(a)
truncated_values = np.trunc(a)
rounded_values = np.round(a)

print(floor_values)
print(ceiling_values)
print(truncated_values)
print(rounded_values)
```
